```python
def sequence_masking(x, mask, value='-inf', axis=None):
    """为序列条件mask的函数
    mask: 形如(batch_size, seq_len)的0-1矩阵；
    value: mask部分要被替换成的值，可以是'-inf'或'inf'；
    axis: 序列所在轴，默认为1；
    除了mask的地方替换，其他地方值不变
    x = torch.randint(1,10,(2,4,3,3))
    mask = torch.Tensor([[1,1,0],[1,0,0]])
    mask2 = torch.unsqueeze(mask, 1)
    mask2 = torch.unsqueeze(mask2, 1)
    """
    if mask is None:
        return x
    else:
        if value == '-inf':
            value = -1e12
        elif value == 'inf':
            value = 1e12
        assert axis > 0, 'greater than 0'
        # attention_mask.unsqueeze(1).unsqueeze(1).expand(batch_size, self.ent_type_size, seq_len, seq_len)

        # 中间维度补
        for _ in range(axis - 1):
            mask = torch.unsqueeze(mask, 1)  # [b,1。。。1，s]
        # 其余维度补
        for _ in range(x.ndim - mask.ndim):
            mask = torch.unsqueeze(mask, mask.ndim)  # [b,1,s,1]
        return x * mask + value * (1 - mask)
```

```python
def apply_rotary_position_embeddings(position_embeddings, *tensors):
    """应用RoPE到tensors中
        其中，sinusoidal.shape=[1, n, d]，tensors为tensor的列表，而
        tensor.shape=[b, n, ..., d]。
    """
    cos_pos = position_embeddings[..., 1::2].repeat_interleave(2, dim=-1)
    sin_pos = position_embeddings[..., ::2].repeat_interleave(2, dim=-1)

    outputs = []
    for tensor in tensors:
        tensor2 = torch.stack([-tensor[..., 1::2], tensor[..., ::2]], -1)
        tensor2 = tensor2.reshape(tensor.shape)
        outputs.append(tensor * cos_pos + tensor2 * sin_pos)
    return outputs[0] if len(outputs) == 1 else outputs
```
